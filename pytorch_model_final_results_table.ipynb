```python
import csv
import pandas as pd
import numpy as np
```

### Merge labels and predictions in one table

```python
df_preds = pd.read_csv('/Users/hela/Code/pata/data_predicted.csv')
data_labeling = pd.read_csv("/Users/hela/Code/pata/data_labeling.csv")  # note that original csv 'data_labeled' is not suitable since does not contain errs
```

```python
print('df_labeling:', len(data_labeling))
print('df_preds:', len(df_preds))
df_labels = data_labeling[data_labeling['label'].notna()]  # note that original csv 'data_labeled' is not suitable since does not contain errs
print('df_labels', len(df_labels))
print('overal:', len(df_preds) + len(df_labels))
```

```python
# making two tables look similar by creating empty columns in one of them:
df_labels['predict'] = np.nan
df_labels['confidence'] = np.nan
```

```python
# combine in one final result table
df_results = df_preds._append(df_labels)  # merged csv (vertical stack)
df_results.sort_values('name', ascending=True, inplace=True)  # sort rows by order
df_results['result'] = df_results['label'].fillna(df_results['predict'])  # create column 'result' with label OR prediction (in label=NaN)
print('rows per random pp:', len(df_results[df_results['name'].str.startswith('02_block')]))  # for manual check
df_results
```

```python
# save:
df_results.to_csv('/Users/hela/Code/pata/data_final_results.csv', index=False)
```
